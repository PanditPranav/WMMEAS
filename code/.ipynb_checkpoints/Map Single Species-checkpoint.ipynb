```python
def geocode_dataframe (df, long, lat):
    from shapely.geometry import Point
    geo_points = df.apply(lambda srs: Point(float(srs[long]), 
                                            float(srs[lat])),
                          axis='columns')
    df_geocoded = gpd.GeoDataFrame(df, geometry=geo_points)
    return df_geocoded
```

```python
cases_geocoded = geocode_dataframe(df= points, long= 'Longitude', lat= 'Latitude')
centers_geocoded = geocode_dataframe(df= centers, long= 'Longitude', lat= 'Latitude')
```
